```python
import pandas as pd
import sys
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer


sys.path.append("../")
```

```python
diamonds = pd.read_csv("../data/train.csv",encoding = "ISO-8859-1")
final_test = pd.read_csv("../data/test.csv",encoding = "ISO-8859-1")
diamonds = diamonds.drop("id", axis=1)
```

The first step is to study the dataset, what kind of information we have

```python
diamonds.dtypes
```

There are 3 categorical columns, with the cut, color and clarity. The data is classified with a code

The correlation between data was then studied

```python
diamonds.corr()
```

We can see that x, y, z and carats have a very strong correlation, in a first iteration they were eliminated, however, the results were better when using "carat" and "x"

In the following plot we can see that there is a relation between carats and price and x and price:

```python
sns.pairplot(diamonds)
```

The next step is to look for null values:

```python
diamonds.isnull().sum().sort_values(ascending=False)
```

```python
diamonds.applymap(lambda x: x == "None").sum().sort_values(ascending=False)
```

The columns "carat" and "x" were kept to train the model:

```python
y = diamonds["price"]
X = diamonds[["carat", "x"]]
X.head()
```

A Pipeline was prepared to scale and test the data, and the mse was calculated, comparing test and train values and their respective predicted prices

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("linear", LinearRegression())
])


pipe.fit(X_train, y_train)


y_pred_train = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)


mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

The mse for the test data was 0.086, not bad for a first iteration

Once the model is prepared, the price of the diamonds in test.csv is predicted and added to a dataset

```python
pipe.fit(X, y)
X_final = final_test[["carat", "x"]]

y_final = pipe.predict(X_final)
final = pd.DataFrame(final_test["id"])
final["price"] = pipe.predict(X_final)
```

```python
final.head()
```

Finally, the dataset is saved in a csv file to submit in Kaggle:

```python
final.to_csv("final.csv", index = False, header = True)
```

### Random forest

To obtain better results, a random forest model was prepared:

```python
y = diamonds["price"]
X = diamonds.drop("price", axis=1)
```

Diamond' cuts can be ordered by quality, Ideal is the best, Fair is the worst cut.

In a first test, the categorical data was transformed with one hot encoding, however, in this particular case, the values in cut, color and clarity need to be sorted by order, as an Ideal cut is much better than a Fair cut. Thus, ordinal encoding was performed:

```python
X["cut"].value_counts(dropna=False)
```

```python
encod = {
    "Fair":4,
    "Good":6,
    "Very Good":8,
    "Premium":9,
    "Ideal":10
}
X["cut"] = X["cut"].apply(lambda x: encod[x])
X.head()
```

```python
X["color"].value_counts(dropna=False)
```

```python
encod = {
    "D":6,
    "E":5,
    "F":4,
    "G":3,
    "H":2,
    "I":1,
    "J":0
}
X["color"] = X["color"].apply(lambda x: encod[x])
X.head()
```

```python
X["clarity"].value_counts(dropna=False)
```

```python
encod = {
    "IF":10,
    "VVS1":9,
    "VVS2":8,
    "VS1":7,
    "VS2":6,
    "SI1":5,
    "SI2":4,
    "I1":3
}
X["clarity"] = X["clarity"].apply(lambda x: encod[x])
X.head()
```

The following code is set as markdown, as I don't need to get dummies anymore, but is kept, in case

```python
#X = pd.get_dummies(X)
```

The model and parameters for the Grid search are prepared as follows. Several iterations were performed to adjust the model and get better mse values

```python
model = RandomForestRegressor()
params = {
    "n_estimators":[200, 210], 
    "max_depth":[36, 38, 40, 42, 44, 46], 
    "min_samples_split":[2], 
    "min_samples_leaf":[2]
}
```

```python
clf = GridSearchCV(estimator=model, param_grid=params, verbose=2, n_jobs = -1)
```

The grid is tested with ALL the values in the dataset, GridSearchCV doesn't need train_test_split

```python
clf.fit(X, y)
```

Once the Grid Search finishes, the best estimator is used and their parameters are obtained.

```python
best = clf.best_estimator_
best
```

```python
clf.best_params_
```

Now we take the best estimators and feed the model with the data, to obtain the mse_train and mse_test, and check the efficiency of the model. The process was repeated several times in order to adjust the estimators to have a better mse. This process can take a long time to do, this is why the "n_jobs = -1" parameter was included in GridSearchCV, it uses the full CPU during the calculations

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
y_pred_train = best.predict(X_train)
y_pred_test = best.predict(X_test)
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)
mse_train, mse_test
```

In order to verify that the results are obtained because the model is good and not by chance, a cross validation score is performed. We can see that the values are good and thus the result is consistent

```python
cross_val_score(estimator=RandomForestRegressor(max_depth=38, min_samples_leaf=2, n_estimators=200), X=X, y=y, cv=10)
```

Finally, we train the model with the data and take the desired columns from test.csv. It's worth noting that is necessary to change the cut, color and clarity columns in the same way we've done before testing the model, in such a way that the data fits the model and is able to predict:

```python
best.fit(X, y)
X_final = final_test[["carat", "cut", "color", "clarity", "depth", "table", "x", "y", "z"]]
```

```python
encod1 = {
    "Fair":4,
    "Good":6,
    "Very Good":8,
    "Premium":9,
    "Ideal":10
}

X_final["cut"] = X_final["cut"].apply(lambda x: encod1[x])

encod2 = {
    "D":6,
    "E":5,
    "F":4,
    "G":3,
    "H":2,
    "I":1,
    "J":0
}

X_final["color"] = X_final["color"].apply(lambda x: encod2[x])

encod3 = {
    "IF":10,
    "VVS1":9,
    "VVS2":8,
    "VS1":7,
    "VS2":6,
    "SI1":5,
    "SI2":4,
    "I1":3
}

X_final["clarity"] = X_final["clarity"].apply(lambda x: encod3[x])

X_final.head()
```

Finally we predict the price of the diamonds and add it to a DataFrame

```python
y_final = best.predict(X_final)
final = pd.DataFrame(final_test["id"])
final["price"] = best.predict(X_final)
```
